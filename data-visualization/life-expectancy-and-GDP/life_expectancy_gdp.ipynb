## Data Vizz

#### GDP throughout the years for each country

```python
# countries GDP
chile_gdp = df.GDP[df.Country == 'Chile']
china_gdp = df.GDP[df.Country == 'China']
germany_gdp = df.GDP[df.Country == 'Germany']
mexico_gdp = df.GDP[df.Country == 'Mexico']
usa_gdp = df.GDP[df.Country == 'United States of America']
zimbabwe_gdp = df.GDP[df.Country == 'Zimbabwe']

# list of countries
countries = ['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe']

# list of years
years = df.Year[df.Country == 'Chile']

ax = plt.axes()
ax.set_facecolor('#2E2E2E')

# plotting gdp against years to see the gdp changes throughout the years
plt.plot(years, chile_gdp, linewidth=2)
plt.plot(years, china_gdp, color='#e71a23')
plt.plot(years, germany_gdp, color='#f6c701', linestyle='dotted', linewidth=2)
plt.plot(years, mexico_gdp, color='#016444', marker='o', linewidth=0.5, markersize=4)
plt.plot(years, usa_gdp, color='#fff')
plt.plot(years, zimbabwe_gdp, linestyle=(0, (3, 5, 1, 5, 1, 5)), linewidth=2)

plt.title('GDP throughout the years for each country')
plt.xlabel('Years')
plt.ylabel('GDP')
plt.legend(countries)
plt.show()
plt.clf()
```

##### Highlights

Above we can see that, in more than 10+ years:
1. Zimbabwe, Mexico and Chile have stable GDP 
2. Germany has a slight increase starting from 2002 to 2008
3. Both China and USA have consistent increasing numbers since 2005

#### Average GDP from 2000 - 2016 for each country

```python
# average gdp for each country
avg_chile_gdp = chile_gdp.mean()
avg_china_gdp = china_gdp.mean()
avg_germany_gdp = germany_gdp.mean()
avg_mexico_gdp = mexico_gdp.mean()
avg_usa_gdp = usa_gdp.mean()
avg_zimbabwe_gdp = zimbabwe_gdp.mean()

# list of gdp averages for each countries
avg_countries_gdp = [avg_chile_gdp, avg_china_gdp, avg_germany_gdp, avg_mexico_gdp, avg_usa_gdp, avg_zimbabwe_gdp]

fig, ax = plt.subplots()
plt.bar(countries, avg_countries_gdp)

plt.title('Average GDP from 2000 - 2016 for each country')
plt.show()
plt.clf()
```

#### Correlation between GDP and Life expectancy

```python
# life expectancy for each country
chile_life = df.Life_expectancy[df.Country == 'Chile']
china_life = df.Life_expectancy[df.Country == 'China']
germany_life = df.Life_expectancy[df.Country == 'Germany']
mexico_life = df.Life_expectancy[df.Country == 'Mexico']
usa_life = df.Life_expectancy[df.Country == 'United States of America']
zimbabwe_life = df.Life_expectancy[df.Country == 'Zimbabwe']

# plotting life expectancy against gdp for each country
# chile
plt.subplot(2, 3, 1, facecolor='#2e2e2e', title='Chile')
plt.scatter(chile_life, chile_gdp, s=8)
plt.ylabel('GDP in Trillions of USD $')

# china
plt.subplot(2,3,2, facecolor='#2e2e2e', title='China')
plt.scatter(china_life, china_gdp, s=8, c='#e71a23')

# germany
plt.subplot(2,3,3, facecolor='#2e2e2e', title='Germany')
plt.scatter(germany_life, germany_gdp, s=8, c='#f6c701')

# mexico
plt.subplot(2,3,4, facecolor='#2e2e2e', title='Mexico')
plt.scatter(mexico_life, mexico_gdp, s=8, c='#016444')
plt.ylabel('GDP in Trillions of USD $')

# usa
plt.subplot(2,3,5, facecolor='#2e2e2e', title='United States of America')
plt.scatter(usa_life, usa_gdp, s=8, c='#fff')
plt.xlabel('Life expectancy at birth (years)')

# zimbabwe
plt.subplot(2,3,6, facecolor='#2e2e2e', title='Zimbabwe')
plt.scatter(zimbabwe_life, zimbabwe_gdp, s=8, c='#fe8114')

plt.subplots_adjust(wspace=0.2, hspace=0.4, top=1.2, right=2)
plt.show()
plt.clf()
```

##### Highlights

All countries have positive correlation between life expectancy and its corresponding GDP. It means that High number of GDP correlates more years of life expectancy. Furthermore, with bare eyes, we can see that Both America & China have linear slope, though China shows steeper slope between 74 and 75 years old.

#### Average Life expectancy

```python
chile_life_avg = chile_life.mean()
china_life_avg = china_life.mean()
germany_life_avg = germany_life.mean()
mexico_life_avg = mexico_life.mean()
usa_life_avg = usa_life.mean()
zimbabwe_life_avg = zimbabwe_life.mean()

avg_countries_life = [chile_life_avg, china_life_avg, germany_life_avg, mexico_life_avg, usa_life_avg, zimbabwe_life_avg]

fig, ax = plt.subplots()
plt.bar(countries, avg_countries_life)

plt.title('Average Life expectancy for each country')
plt.show()
plt.clf()
```

#### The distribution of Life expectancy

```python

# chile
plt.subplot(2, 3, 1, facecolor='#2e2e2e', title='Chile')
plt.hist(chile_life)
# plt.ylabel('GDP in Trillions of USD $')

# china
plt.subplot(2,3,2, facecolor='#2e2e2e', title='China')
plt.hist(china_life, facecolor='#e71a23')

# germany
plt.subplot(2,3,3, facecolor='#2e2e2e', title='Germany')
plt.hist(germany_life, facecolor='#f6c701')

# mexico
plt.subplot(2,3,4, facecolor='#2e2e2e', title='Mexico')
plt.hist(mexico_life, facecolor='#016444')
# plt.ylabel('GDP in Trillions of USD $')

# usa
plt.subplot(2,3,5, facecolor='#2e2e2e', title='United States of America')
plt.hist(usa_life, facecolor='#fff')
plt.xlabel('Distribution of life expectancy at birth (years) for each country', size='xx-large')

# zimbabwe
plt.subplot(2,3,6, facecolor='#2e2e2e', title='Zimbabwe')
plt.hist(zimbabwe_life, facecolor='#fe8114')


plt.subplots_adjust(wspace=0.2, hspace=0.4, top=1.2, right=2)
plt.show()
plt.clf()
```

##### Highlights

Chile has normal distribution on its life expectancy, meanwhile China and Zimbabwe have skewed distribution. 

1. China has more people who live until 75 years old than people who live before or after 75 years old. 
2. In Zimbabwe, it's more common to see people who live until 45 years old, rather than people who live after that age. They also have the lowest life expectancy. 
3. In Germany, people have the highest life expectancy up to 81 years old, than any other country, with 78 years old as the lowest.
